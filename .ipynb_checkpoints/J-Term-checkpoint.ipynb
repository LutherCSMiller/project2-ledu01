# Question: "How do the countries Luther visited for J-Term distribute on the world map?"

## I used the J-Term data provided by the Center for Global Learning (dating back to 1991), count the visited time for each country and plot it on the world map using plotly. 

```python
df = pd.read_excel("J-Term.xlsx")
```

```python
d = df[['Year', 'Country #1', 'Country #2', 'Country #3', 'Country #4', 'Country #5']]
```

```python
d.columns = ['year', 'country1', 'country2', 'country3', 'country4', 'country5']
d=d.fillna('')
```

```python
d.country1 = d.country1.apply(lambda x: x.rstrip())
d.country2 = d.country2.apply(lambda x: x.rstrip())
d.country3 = d.country3.apply(lambda x: x.rstrip())
d.country4 = d.country4.apply(lambda x: x.rstrip())
d.country5 = d.country5.apply(lambda x: x.rstrip())
```

```python
masterlist = []
for i in ['country1', 'country2', 'country3', 'country4', 'country5']:
    for j in sorted(d[i].unique()):
        if j not in masterlist:
            masterlist.append(j)
```

```python
ranking = []
yearlookup = []
country = []
datatable = {}
```

```python
for x in masterlist:
    if x != '':    
        country.append(x)
        ranking.append(d[(d.country1 == x)|(d.country2 == x)|(d.country3 == x)|(d.country4 == x)|(d.country5 == x)].year.count())
        yearlookup.append(sorted(d[(d.country1 == x)|(d.country2 == x)|(d.country3 == x)|(d.country4 == x)|(d.country5 == x)].year.unique()))
datatable['Country'] = country
datatable['Ranking'] = ranking
datatable['Year'] = yearlookup
```

```python
df = pd.DataFrame(datatable)
```

```python
df = df.sort_values('Ranking', ascending = False).reset_index(drop=True)
```

```python
df2 = pd.read_excel("country_code_web.xls")
```

```python
df2
```

```python
def matchcountrycode(country):
    if country in ['England', 'Scotland', 'Northern Ireland']:
        return 'GBR'
    elif country == 'Leeward Islands':
        return 'NLD'
    elif country == 'Canary Islands': 
        return 'ESP'
    else:
        if df2[df2['Country'] == country].Code.get_values().size == 0:
            return ''
        else:
            return df2[df2['Country'] == country].Code.get_values()[0]
```

```python
df['Code'] = df.Country.apply(lambda x: matchcountrycode(x))
df['Year'] = df['Year'].apply(lambda x: str(x)[1:-1])
df['Text'] = df['Country'] + '<br>' + "Visited in " + df['Year']
```

```python
df
```

```python
data = [ dict(
        type = 'choropleth',
        locations = df['Code'],
        z = df['Ranking'],
        text = df['Text'],
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            tickprefix = '',
            title = 'Visited times'),
      ) ]

layout = dict(
    title = 'J-Term Countries',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Miller'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='d3-world-map' )
```
